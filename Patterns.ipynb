🟢 LEVEL 1: Basic Warm-up (Very Easy)



1. Square Pattern



````python
class Solution:
  def pattern1(self, n):
    for i in range(n):
      print('*' * n)

obj = Solution()
obj.pattern1(4)
````

````python
class Solution:
  def pattern2(self, n):
    for i in range(1,n+1):
      print('*'*i)


obj = Solution()
obj.pattern2(4)
````

11. Pyramid




```

  ***
 *****

```



````python
class Solution:
  def pattern11(self, n):
    for i in range(n):
      spaces = n-i-1
      stars = 2*i + 1
      print(' '* spaces + '*'*stars )

obj = Solution()
obj.pattern11(4)
````

13. Diamond ⭐⭐⭐




```
   *
  ***
 *****
*******
 *****
  ***
   *
```



````python
class Solution:
  def pattern13(self, n):
    for i in range(1,n+1):
      spaces = n - i
      stars = 2*i - 1
      print(' '* spaces + '*'*stars )
    for i in range(1, n):
      spaces = i
      stars = 2*(n-i)-1
      print(' '* spaces + '*'*stars )


obj = Solution()
obj.pattern13(4)
````

14. Hollow Pyramid ⭐⭐⭐



```
   *
  * *
 *   *
*******
```



````python
class Solution:
  def pattern14(self, n):
    for i in range(1,n+1):
      for j in range(n-i):
        print(' ', end='')
      for j in range(2*i-1):
        if i == 1 or j ==0 or i ==n or j == 2*i-2:
          print('*', end='')
        else:
          print(' ', end='')


      print()




obj = Solution()
obj.pattern14(4)
````

```

***  ***
**    **
*      *
```



````python
class Solution:
  def pattern17(self, n):
    for i in range(1,n+1):
      for j in range(i):
        print('*', end='')
      for j in range(2*(n-i)):
        print(' ', end='')
      for j in range(i):
        print('*', end='')
      print()
    for i in range(n,0,-1):
      for j in range(i):
        print('*', end='')
      for j in range(2*(n-i)):
        print(' ', end='')
      for j in range(i):
        print('*', end='')
      print()






obj = Solution()
obj.pattern17(4)
````

18. Hollow Diamond ⭐⭐⭐⭐





```
   *
  * *
 *   *
  * *
   *
```





````python
class Solution:
  def pattern18(self, n):
    for i in range(0,n):
      for j in range(1,n-i):
        print(' ', end='')
      for j in range(2*i+1):
        if i == 0 or j ==0 or j == 2*i:
          print('*', end= '')
        else:
          print(' ', end='')
      print()
    for i in range(n-1,0, -1):
      for j in range(1, n-i+1):
        print(' ', end='')
      for j in range(1,2*i):
        if i ==1 or j == 1 or j == 2*i -1:
          print('*', end= '')
        else:
          print(' ', end= '')

      print()





obj = Solution()
obj.pattern18(4)
````

20. Spiral Matrix ⭐⭐⭐⭐⭐



```
1 2 3
8 9 4
7 6 5
```



````python
class Solution:
  def pattern20(self, n):
    matrix = [[0]*n for  _ in range(n)]
    num = 1
    top , bottom = 0, n-1
    left , right = 0, n-1

    while top <= bottom and left <= right:


    # left to right
      for j in range(left , right+1):
        matrix[top][j] = num
        num += 1
      top += 1

    # top to bottom
      for i in range(top , bottom+1):
        matrix[i][right] = num
        num += 1
      right -= 1

    # right to left
      if top <= bottom :
        for j in range(right , left-1, -1):
          matrix[bottom][j] = num
          num += 1
      bottom -= 1

    # bottom to top
      if left  <= right :
        for i in range(bottom , top-1 , -1):
          matrix[i][left] = num
          num += 1
      left += 1






    # print matrix
    for row in matrix:
      print(*row)
obj = Solution()
obj.pattern20(3)
````
